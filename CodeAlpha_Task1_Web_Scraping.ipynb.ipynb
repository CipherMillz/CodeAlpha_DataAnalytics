```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("✅ All libraries imported successfully!")
```

```python
url = "https://quotes.toscrape.com/"
response = requests.get(url)

print(response.status_code)   # Should print 200 = Success
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
quotes = []
authors = []

for quote in soup.find_all("div", class_="quote"):
    text = quote.find("span", class_="text").get_text()
    author = quote.find("small", class_="author").get_text()
    
    quotes.append(text)
    authors.append(author)

print("Total quotes scraped:", len(quotes))
print("\nFirst 3 quotes:")
for i in range(3):
    print(f"{i+1}. {quotes[i]} - {authors[i]}")
```

```python
# Create a nice table from the scraped data
df = pd.DataFrame({
    "Quote": quotes,
    "Author": authors
})

# Show the first 5 rows
df.head()
```

```python
# Save it so we can use it later
df.to_csv("quotes.csv", index=False)
print("✅ File saved as quotes.csv")
```

```python
print("Total quotes:", len(df))
print("\nMost common authors:")
print(df['Author'].value_counts().head(5))

# Show all quotes by Albert Einstein
einstein_quotes = df[df['Author'] == "Albert Einstein"]
print("\nAlbert Einstein quotes:")
print(einstein_quotes)
```
